```python
import pandas as pd
import numpy as np
import einsteinds.utils as utils
import einsteinds.db as edb
from pandas.io.json import json_normalize
import json
import datetime
import sklearn

# load the database credentials from file
with open('../creds/local_creds.json') as json_data:
    creds = json.load(json_data)

# set up a database with credentials
db = edb.Database(creds)

# get the unique combinations of eventCategory, eventAction and eventLabel
cal = list(db._client['production']['eventCollection'].aggregate([{
    '$group': {
        '_id': {
            'eventCategory': "$eventCategory", 
            'eventAction': "$eventAction", 
            'eventLabel': "$eventLabel"
        }
    }
}]))

unique_event_types = json_normalize(cal)

unique_event_types.head()
unique_event_types = unique_event_types[unique_event_types['_id.eventCategory'] != 'session'].reset_index(drop=True).sort_values(by=['_id.eventCategory','_id.eventAction','_id.eventLabel'])
unique_event_types.columns = ['eventAction','eventCategory','eventLabel']
unique_event_types = unique_event_types[['eventCategory','eventAction','eventLabel']]
unique_event_types.reset_index(drop=True, inplace=True)
unique_event_types = unique_event_types[unique_event_types.eventCategory != 'category']
unique_event_types.loc[unique_event_types.eventLabel == 'bitcoin', 'eventLabel'] = 'BTC'

with pd.option_context("display.max_rows",500):
    display(unique_event_types.sort_values(by=['eventCategory','eventAction','eventLabel']).set_index(['eventCategory','eventAction','eventLabel']))
```
